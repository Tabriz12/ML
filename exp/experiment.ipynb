```python
import pandas as pd

from ml.error import MSE
from ml.classic_ml.models import LinearRegression

data = pd.read_csv(
    "../data/Student_Marks.csv",
)

X: pd.DataFrame = data[["number_courses", "time_study"]]

Y: pd.Series = data["Marks"]

x_train, y_train = X.loc[:90], Y.loc[:90]
x_test, y_test = X.loc[90:], Y.loc[90:]

linear = LinearRegression()

linear.fit(x_train, y_train, 5000, lr=1e-3)

print(linear.B, linear.W)

predictions = linear.predict(x_test)

print(MSE(y_test, predictions))
```

```python
import pandas as pd

from ml.classic_ml.models import LogisticRegression

from ml.error import confusion_matrix

data = pd.read_csv("../data/Social_Network_Ads.csv")


X: pd.DataFrame = data.loc[:, ~data.columns.isin(["User ID", "Purchased"])]

Y: pd.Series = data["Purchased"]


X["Gender"] = X["Gender"].map({"Male": 0, "Female": 1})

X["Age"] = (X["Age"] - X["Age"].min()) / (X["Age"].max() - X["Age"].min())

X["EstimatedSalary"] = (X["EstimatedSalary"] - X["EstimatedSalary"].min()) / (
    X["EstimatedSalary"].max() - X["EstimatedSalary"].min()
)


split_index = int(0.9 * len(X))

x_train, y_train = X.loc[:split_index], Y.loc[:split_index]
x_test, y_test = X.loc[split_index:], Y.loc[split_index:]

print(f"x_train shape: {x_train.shape}, y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape}, y_test shape: {y_test.shape}")

lg = LogisticRegression()

lg.fit(x_train, y_train, epochs=5000, lr=0.05)


y_pred, y_prob = lg.predict(x_test)

print(y_pred)
print(y_prob)


print(confusion_matrix(y_test, y_pred))
```
